```python
import pandas as pd
import os
print("1. Setting up directories...")
os.makedirs("data/processed", exist_ok=True)
def extract_prompt_response_pairs(df):
    prompts = []
    responses = []
    for dialog in df['dialog']:
        if len(dialog) > 1:
            for i in range(len(dialog) - 1):
                prompts.append(dialog[i].strip())
                responses.append(dialog[i+1].strip())  
    return pd.DataFrame({
        'prompt': prompts,
        'response': responses
    })
print("\n2. Loading Raw Parquet Files...")
raw_train = pd.read_parquet("data/raw/train-00000-of-00001.parquet")
raw_val = pd.read_parquet("data/raw/validation-00000-of-00001.parquet")
raw_test = pd.read_parquet("data/raw/test-00000-of-00001.parquet")
print("\n3. Processing and Flattening Data...")
processed_train = extract_prompt_response_pairs(raw_train)
processed_val = extract_prompt_response_pairs(raw_val)
processed_test = extract_prompt_response_pairs(raw_test)
print(f"Processed Train Pairs: {len(processed_train)}")
print(f"Processed Validation Pairs: {len(processed_val)}")
print(f"Processed Test Pairs: {len(processed_test)}")
print("\n4. Saving to data/processed/ ...")
processed_train.to_csv("data/processed/train_pairs.csv", index=False)
processed_val.to_csv("data/processed/val_pairs.csv", index=False)
processed_test.to_csv("data/processed/test_pairs.csv", index=False)
print("\nData preprocessing complete! Files are ready in data/processed/")  
```
